```python
import pandas as pd
import matplotlib.pyplot as plt
sets=pd.read_csv('Data-Analytics-Projects/Data/sets.csv')
themes_by_year=sets.groupby('year').agg({'theme_id':pd.Series.nunique})
themes_by_year.rename(columns={'theme_id':'unique_themes'},inplace=True)
plt.figure(figsize=(20,15))
plt.xlabel("year",fontsize=12)
plt.ylabel("no of themes",fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylim(0,100)
plt.plot(themes_by_year.index[:-1],themes_by_year.unique_themes[:-1])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
sets=pd.read_csv('Data-Analytics-Projects/Data/sets.csv')
sets_by_year=sets.groupby('year').count()
themes_by_year=sets.groupby('year').agg({'theme_id':pd.Series.nunique})
themes_by_year.rename(columns={'theme_id':'unique_themes'},inplace=True)
ax1=plt.gca()
ax2=ax1.twinx()
ax1.set_xlabel('year')
ax1.set_ylabel('number of sets',color='green')
ax2.set_ylabel('number of themes',color='blue')
ax1.plot(sets_by_year.index[:-1],sets_by_year.set_num[:-1],'g')
ax2.plot(themes_by_year.index[:-1],themes_by_year.unique_themes[:-1],'b')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
sets=pd.read_csv('Data-Analytics-Projects/Data/sets.csv')
parts_per_set=sets.groupby('year').agg({'num_parts':pd.Series.mean})
plt.scatter(parts_per_set.index[:-1],parts_per_set.num_parts[:-1])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
sets=pd.read_csv('Data-Analytics-Projects/Data/sets.csv')
sets=sets['theme_id'].value_counts()
new_sets=pd.DataFrame({'id':sets.index,'set_count':sets.values})
new_sets.head()
themes=pd.read_csv('Data-Analytics-Projects/Data/themes.csv')
merged=pd.merge(new_sets,themes,on='id')

plt.figure(figsize=(14,8))
plt.xticks(fontsize=14, rotation=45)
plt.yticks(fontsize=14)
plt.ylabel('Nr of Sets', fontsize=14)
plt.xlabel('Theme Name', fontsize=14)

plt.bar(merged.name[:10], merged.set_count[:10])
```
