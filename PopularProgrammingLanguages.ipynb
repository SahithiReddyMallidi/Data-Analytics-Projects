```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("Data-Analytics-Projects/Data/ProgrammingLanguages_tags.csv",names=['Date','Tag','Posts'],header=0)
df['Date']=pd.to_datetime(df['Date'])
new_df=df.pivot(columns='Tag',index='Date',values='Posts');
new_df.fillna(0,inplace=True)
roll_df=new_df.rolling(window=45).mean()
plt.figure(figsize=(20,15))
plt.xlabel('Date', fontsize=12)
plt.ylabel('No of Posts', fontsize=12)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.ylim(0,30000)
for col in roll_df.columns:
  plt.plot(roll_df.index,roll_df[col],label=roll_df[col].name,linewidth=3)
plt.legend(fontsize=15)
```
